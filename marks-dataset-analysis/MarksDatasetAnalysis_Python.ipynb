# Assignment - Numpy & Pandas

```python
import numpy as np
import pandas as pd

#Create Student IDs (1 to 100)
student_ids = np.arange(1, 101)

#Generate random marks between 30 and 100
np.random.seed(42)  # For reproducible results
marks = np.random.randint(30, 101, size=(100, 5))

#Create DataFrame
columns = ["Math", "Science", "English", "History", "Computer"]
df = pd.DataFrame(marks, columns=columns)
df["StudentID"] = student_ids

#Move StudentID to the front
df = df[["StudentID"] + columns]

#Add Average column
df["Average"] = df[columns].mean(axis=1)

#Add Grade column based on conditions
def assign_grade(avg):
    if avg >= 85:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "D"

df["Grade"] = df["Average"].apply(assign_grade)

#Display final DataFrame
print(df)



import matplotlib.pyplot as plt

columns = ["Math", "Science", "English", "History", "Computer"]

#Grade Distribution (Bar Chart)
plt.figure(figsize=(8,5))
df["Grade"].value_counts().sort_index().plot(kind="bar")
plt.title("Distribution of Grades")
plt.xlabel("Grade")
plt.ylabel("Count")
plt.show()   # <<< SHOW PLOT

#Boxplot of All Subjects
plt.figure(figsize=(10,6))
df[columns].boxplot()
plt.title("Boxplot of Subject Scores")
plt.ylabel("Marks")
plt.show()   # <<< SHOW PLOT

#Correlation Heatmap
plt.figure(figsize=(7,6))
corr = df[columns].corr()
plt.imshow(corr)
plt.xticks(range(len(columns)), columns, rotation=45)
plt.yticks(range(len(columns)), columns)
plt.colorbar()
plt.title("Correlation Heatmap")
plt.show()   #SHOW PLOT


print("Plots saved successfully!")

```
